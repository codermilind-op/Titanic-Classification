```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv('train.csv')
```

```python
import seaborn as sns
numeric_columns = titanic_data.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_columns.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2)
for train_indices, test_indices in split.split(titanic_data, titanic_data[["Survived", "Pclass","Sex"]]):
    strat_train_set = titanic_data.loc[train_indices]
    strat_test_set = titanic_data.loc[test_indices]
```

```python
plt.subplot(1,2,1)
strat_train_set['Survived'].hist()
strat_train_set['Pclass'].hist()
plt.subplot(1,2,2)
strat_test_set['Survived'].hist()
strat_test_set['Pclass'].hist()

plt.show()
```

```python
strat_train_set.info()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.impute import SimpleImputer
class AgeImputer(BaseEstimator, TransformerMixin):
    def fit(self,X, y=None):
        return self
    def transform(self, X):
        imputer = SimpleImputer(strategy="mean")
        X['Age'] = imputer.fit_transform(X[['Age']])
        return X
    
```

```python
from sklearn.preprocessing import OneHotEncoder
class FeatureEncoder(BaseEstimator,TransformerMixin):
    def fit(self, X, Y=None):
        return self
    def transform(self, X):
        encoder = OneHotEncoder()
        matrix = encoder.fit_transform(X[['Embarked']]).toarray()

        column_names = ["C","S","Q","N"]
        for i in range(len(matrix.T)):
            X[column_names[i]] = matrix.T[i]
        matrix = encoder.fit_transform(X[['Sex']]).toarray()
        column_names = ["Female","Male"]
        for i in range(len(matrix.T)):
            X[column_names[i]] = matrix.T[i]
        return X    
```

```python
class FeatureDropper(BaseEstimator, TransformerMixin):
    def fit(self, X ,y=None):
        return self
    def transform(self,X):
        return X.drop(["Embarked","Name","Ticket","Cabin","Sex","N"],axis=1,errors="ignore")
```

```python
from sklearn.pipeline import Pipeline
pipeline=Pipeline([("ageimputer", AgeImputer()),
                   ("featureencoder", FeatureEncoder()),
                   ("featuredropper", FeatureDropper())])
```

```python
strat_train_set=pipeline.fit_transform(strat_train_set)
```

```python
strat_train_set
```

```python
strat_train_set.info()
```

```python
from sklearn.preprocessing import StandardScaler
X = strat_train_set.drop(['Survived'], axis=1)
Y = strat_train_set['Survived']
scaler = StandardScaler()
X_data = scaler.fit_transform(X)
Y_data = Y.to_numpy()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

clf = RandomForestClassifier()
param_gird = [
    {"n_estimators":[10,100,200,500], "max_depth": [None,5,10], "min_samples_split": [2,3,4]
    }]
grid_search = GridSearchCV(clf, param_gird, cv=3, scoring="accuracy", return_train_score=True)
grid_search.fit(X_data, Y_data)
```

```python
final_clf = grid_search.best_estimator_
```

```python
final_clf
```

```python
strat_test_set = pipeline.fit_transform(strat_test_set)
```

```python
X_test = strat_test_set.drop(['Survived'], axis=1)
Y_test = strat_test_set['Survived']

scaler = StandardScaler()
X_data_test = scaler.fit_transform(X_test)
Y_data_test = Y_test.to_numpy()
```

```python
final_clf.score(X_data_test, Y_data_test)
```

```python
final_data = pipeline.fit_transform(titanic_data)
```

```python
final_data
```

```python
X_final = final_data.drop(['Survived'], axis=1)
Y_final = final_data['Survived']

scaler = StandardScaler()
X_data_final = scaler.fit_transform(X_final)
Y_data_final = Y_final.to_numpy()
```

```python
prod_clf = RandomForestClassifier()
param_gird = [
    {"n_estimators":[10,100,200,500], "max_depth": [None,5,10], "min_samples_split": [2,3,4]
    }]
grid_search = GridSearchCV(prod_clf, param_gird, cv=3, scoring="accuracy", return_train_score=True)
grid_search.fit(X_data_final, Y_data_final)
```

```python
prod_final_clf = grid_search.best_estimator_
```

```python
titanic_test_data = pd.read_csv("test.csv")
```

```python
final_test_data = pipeline.fit_transform(titanic_test_data)
```

```python
X_final_test = final_test_data
# X_final_test = X_final_test.ffill
X_final_test = final_test_data.ffill()
scaler = StandardScaler()
X_data_final_test = scaler.fit_transform(X_final_test)
```

```python
predictions = prod_final_clf.predict(X_data_final_test)
```

```python
final_df = pd.DataFrame(titanic_test_data['PassengerId'])
final_df['Survived'] = predictions
final_df.to_csv("predictions.csv", index=False)
```
